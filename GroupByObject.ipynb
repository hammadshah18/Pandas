<h1>find ranking of each movie in the group according to IMDB score</h1>

```python
def movie_rank(group):
    group['Genre_Rank']=group['IMDB_Rating'].rank(ascending=False)
    return group
```

```python
gb.apply(movie_rank)
```

<h1>Find the Normalizad IMDB rating group wise</h1>

```python
def normal(group):
    group['Normalized']=((group['IMDB_Rating']-group['IMDB_Rating'].min())/(group['IMDB_Rating'].max()-group['IMDB_Rating'].min()))
    return group
```

```python
gb.apply(normal)
```
