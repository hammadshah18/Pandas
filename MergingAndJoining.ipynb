<h1># IPL Problems

# find top 3 studiums with highest sixes/match ratio
# find orange cap holder of all the seasons</h1>

```python
match.head(1)
```

```python
dlv.head(1)
```

```python
comp=match.merge(dlv,how='inner',left_on='id',right_on='match_id')
```

```python
comp.head(1)
```

```python
sixes=comp[comp['batsman_runs']==6]
sixes.head(1)
```

```python
num_sixes=sixes.groupby('venue')['venue'].count()
```

```python
num_match=match['venue'].value_counts()
```

```python
(num_sixes/num_match).sort_values(ascending=False).head(3)
```

</h1>

```python
comp.groupby(['season','batsman'])['batsman_runs'].sum().reset_index().sort_values('batsman_runs',ascending=False).drop_duplicates(subset=['season'],keep='first').sort_values('season')
```
