```python
import pandas as pd

# Sample DataFrame
data = {
    'ID': [1, 2, 3, 4, 5, 6],
    'Keyword': ['apple', 'banana', 'apple', 'orange', 'banana', 'grape']
}

df = pd.DataFrame(data)

# Detect duplicate keywords
duplicates = df[df.duplicated(subset=['Keyword'], keep=False)]

print("Duplicates:")
print(duplicates)

# Remove duplicates and keep the first occurrence
df_no_duplicates = df.drop_duplicates(subset=['Keyword'], keep='first')

print("\nDataFrame after removing duplicates:")
print(df_no_duplicates)

# Count occurrences of each keyword
keyword_counts = df['Keyword'].value_counts()

print("\nKeyword occurrences:")
print(keyword_counts)
```

```python
# find euclidean distance with Apply keywords

dis=pd.DataFrame({'1st Point':[(2,3),(4,5),(-3,9),(-1,7)],
                  '2nd Point':[(4,3),(9,2),(0,9),(8,1)]})

dis
```

```python
def euclidean(row):
    pt_A=row['1st Point']
    pt_B=row['2nd Point']
    return ((pt_A[0]-pt_B[1])**2 + (pt_A[1]-pt_B[0])**2)**0.5
    
```

```python
dis.apply(euclidean,axis=1)
```
