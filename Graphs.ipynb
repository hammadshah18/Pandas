```python
rec=pd.read_csv('batsman_season_record.csv')
rec
```

```python
plt.bar(rec['batsman'],rec['2015'],width=0.4,label='2015')
plt.bar(rec['batsman'],rec['2016'],width=0.2,label='2016')
plt.bar(rec['batsman'],rec['2017'],width=0.1,label='2017')
plt.legend()
plt.title('Record of every Batsman in ipl in different Seasons')
plt.xlabel('Players Name')
plt.ylabel('Players Runs')
```

```python
plt.bar(np.arange(rec.shape[0])- 0.2,rec['2015'],width=0.2)
plt.bar(np.arange(rec.shape[0]),rec['2016'],width=0.2)
plt.bar(np.arange(rec.shape[0])+ 0.2,rec['2017'],width=0.2)
plt.xticks(np.arange(rec.shape[0]),rec['batsman'])
plt.grid()
plt.show()
```

```python
# problem
children = [10,20,40,10,30]
colors = ['redredredredredredredredred','blueredredredredredredred','greenredredredredred','yellow','pink']

plt.bar(colors,children,color='black')
plt.xticks(rotation='vertical')
plt.show()
```

<h1>Stacked Bar Chart</h1>

```python
plt.bar(rec['batsman'],rec['2017'],label='2017')
plt.bar(rec['batsman'],rec['2016'],bottom=rec['2017'],label='2016')
plt.bar(rec['batsman'],rec['2015'],bottom=(rec['2016'] +rec['2017']),label='2015')
plt.grid()

plt.legend()
```

<h1>Pie chart</h1>

Univariate/Bivariate Analysis


Categorical vs numerical


Use case - To find contibution on a standard scale

```python
data=[12,34,54,78,90]
sub=['math','urdu','science','pst','comp']
plt.pie(data,labels=sub,autopct='%0.1f%%',explode=[0.1,0,0,0.1,0],shadow=True)
plt.show()
```
